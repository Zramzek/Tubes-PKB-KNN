```python
new_data = my_data.drop('CondensedBinary2DGeometry', axis=1)
new_data
```

```python
user_data = [200,55]
columns = list(new_data)
columns
```

```python
data_arr = []
dist_arr = []

for i in range (0,200):
    for j in columns:
        print(new_data[j][i])
        elem = new_data[j][i]
        data_arr.append(elem)
    print(data_arr)
    dist = ((user_data[0]-data_arr[0]) **2 + (user_data[1]-data_arr[1]) **2) **0.5
    
    dist_arr.append(dist)
    data_arr = []
```
